12. Después de que te preguntaran por los animes que tienen mejor score, se te ocurre crear una columna que indique cuales animes son "buenos". Para esto, el área de producto te indica que un "mal" anime tiene score bajo 4.0, uno "intermedio" tiene score entre 4.0 y 8.0, y uno "bueno" tiene score sobre 8.0. **Usa el método apply para crear una nueva columna "quality" que tenga como valor "Malo", "Intermedio" o "Bueno" según la definición anterior. Luego, muestra los 5 primeros registros**

```python
import numpy as np

new_df["quality"] = pd.cut(
    new_df["score"],
    bins=[-np.inf, 4.0, 8.0, np.inf], #separaciones para las categorias
    labels=["Malo", "Intermedio", "Bueno"]
)

new_df.head(5)
```

# Visualicemos los datos

17. Muestre el histograma de todas las puntuaciones de animes

```python
plt.figure(figsize=(10,6))
plt.hist(new_df['score'], bins=20, edgecolor='black', alpha=0.7)
plt.title('Puntuaciones de animes')
plt.xlabel('Puntuación')
plt.ylabel('Frecuencia')
plt.show()
```

18. Muestre la densidad de puntajes para los principales tipos de anime (TV, OVA, Movie)

```python
import seaborn as sns

df_main_types = new_df[new_df['type'].isin(['TV', 'OVA', 'Movie'])]

plt.figure(figsize=(10, 6))
sns.kdeplot(data=df_main_types, x='score', hue='type', fill=True, alpha=0.5, common_norm=False)

plt.xlabel('Puntuación')
plt.ylabel('Densidad')
plt.title('Densidad de puntajes para los principales tipos de anime')

plt.show()
```

19. ¿Cuál tipo de anime es más exitoso? ¿Se está tomando en cuenta la cantidad de animes por tipo al afirmar eso?

R: De primeras, pareciera que el tipo de anime más exitoso es el de TV, aunque al ser el tipo más predominante esto da una imagen falsa del tipo más exitoso. Luego, al agregar common_norm=False al gráfico se está normalizando cada curva independientemente para una comparativa más justa, mostrando así una distribución de densidad bastante similar entre los 3 tipos más predominantes (un poco más bajo en el caso de los OVAS)
